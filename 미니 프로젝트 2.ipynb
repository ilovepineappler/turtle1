```python
#로딩
from pandas import Series,DataFrame
import numpy as np
import pandas as pd
import os
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt

encoding ='latin1'

unames =['user_id','gender','age','occupation','zip']
rnames =['user_id','movie_id','rating','timestamp']
mnames =['movie_id','title','genres']

users = pd.read_csv('C:/Users/edu/Desktop/ml-1m/users.dat',sep='::',header=None,names=unames,encoding=encoding ,engine='python')
ratings = pd.read_csv('C:/Users/edu/Desktop/ml-1m/ratings.dat',sep='::',header=None,names=rnames,encoding=encoding , engine='python')
movies= pd.read_csv('C:/Users/edu/Desktop/ml-1m/movies.dat',sep='::',header=None,names=mnames,encoding=encoding ,engine='python')
```

```python

# 연령별 관람 순위 (어느 연령대가 가장 영화를 많이 보는가)

data=pd.merge(pd.merge(ratings,users),movies)
display(data)
age=data.age.unique()
display(age)
ages=sorted(age)
display(ages)
bins=[0,10,20,30,40,50,60]
data['age_range'] = pd.cut(data.age, bins, right=False)
age_ranges=data.groupby(data['age_range']).count()
display(age_ranges)
age_=age_ranges.age_range.rank()
age_
```
